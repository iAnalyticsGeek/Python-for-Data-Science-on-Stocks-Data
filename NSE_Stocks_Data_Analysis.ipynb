```python
# load the packages
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# load the list of tickers listed at NSE from the csv file
list_tickers_nse = pd.read_csv("all_stock_codes_nse.csv")
```

```python
list_tickers_nse.head(3)
```

```python
# get the tickers data into a list
yahoo_symbols = list_tickers_nse.Yahoo_Symbol.to_list()
```

```python
# type of the data
type(yahoo_symbols)
```

```python
# first three elements in the list
print(yahoo_symbols[0:3])
```

```python
# what is the length of list elements?
print(len(yahoo_symbols))
```

```python
# get the data for these tickers from yahoo finance, default is one day interval data
tickers_data = yf.download(yahoo_symbols, start="2020-01-01", end="2020-04-10", group_by="ticker")
```

```python
tickers_data.head(3)
```

```python
# to access individual ticker from the dataframe
tickers_data['GNA.NS'][['Open', 'Close']].head()
```

```python
# save the data for future use
tickers_data.to_csv("NSE_Tickers_Data.csv")
```

```python
# make a copy of full dataset
closed_price_df = tickers_data.copy()
```

```python
# clean the dataset into a dataset that contains only 'Close' prices, datewise, for all the scrips
closed_price_df = closed_price_df.iloc[:, closed_price_df.columns.get_level_values(1)=='Close']
closed_price_df.head(3)
```

```python
# drop the level 'Close'
closed_price_df.columns = closed_price_df.columns.droplevel(1)
closed_price_df.head(3)
```
